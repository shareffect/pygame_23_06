```python
import pygame

# 초기화

pygame.init()
# (5,0)  성공 5, 실패 0 이라는 의미인데, pygame 만든 사람이 뭐 ( , ) 의미를 정해 놨겠죠.

# 2. 게임화면 설정
size = [400,900]
screen = pygame.display.set_mode(size)
title = "Using PyGame"
pygame.display.set_caption(title)


# 3. 게임 내부 설정 > 변수 

clock = pygame.time.Clock()
black = (0,0,0) # 검정색임.# 숫자범위. 0 ~ 255 까지. # 헥사코드도 있습니다. 
player = pygame.image.load("C:/Users/Administrator/Pictures/pygame/player.png").convert_alpha()


# ext = player.get_rect()[2:4]
# size = 0.3

# player = pygame.transform.scale(
#     player,
#     (int(ext[0]*size), int(ext[1]*size))
# )



k = 0


player = pygame.transform.scale(player,(50,80))
player_width, player_height = player.get_size()
player_x = round((size[0]-player_width)/2)
player_y = size[1] - player_height - 50


# 4. 메인 이벤트 ( 프로그램에서 중요한 로직이 들어가는 부분 )
# while(True)  # 게임을 끄기 전까지는 계속 반복할 것이기 때문에.
# ESC 라면 반복문 탈출. 탈출은 break

# 무한반복하는 와중에, 키 감지하면 감지한 대로 반응하고 계속 반복은 하는 것.

system_exit = 0

while(system_exit == 0):  # 나중에 탈출하고 싶으면 system_exit 를 0이 아닌 다른 값으로 바꾸면 됨.
    #FPS(Frame Per Second) 
    clock.tick(10)
    # 사람의 눈. 180 ~ 244 프레임 정도 됩니다. 
    
    # 입력 감지
    for event in pygame.event.get():  # 여러가지 키 기능을 동시에 처리할 수 있도록 해 주어야 합니다. 
        if event.type == pygame.QUIT:
            system_exit = 1  # 만약 break 를 넣었으면 바로 멈췄겠죠. while 에서 참 거짓 판단하는 과정 없이.
            # 여기서는 안전성을 위해서 바로 break 안하고 system_exit 썼습니다.
            
    # 변화
    k += 1
            
    # 전사작업
    screen.fill(black)
    screen.blit(player, (player_x,player_y))
    # screen.blit(player, (120,700))
    
    
    # 업데이트
    pygame.display.flip()
    
    
# 종료
pygame.quit()
```
